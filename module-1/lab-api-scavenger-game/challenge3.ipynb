```python
import json
import requests
import os
from dotenv import load_dotenv
import re
load_dotenv()
url = "https://api.github.com/users/gthecrack"
res = requests.get(url)
print(res)
def getFromGithub(endpoint, queryParams={}):
    url = f"https://api.github.com{endpoint}"

    apiKey = os.getenv("GITHUB_APIKEY")
    if not apiKey:
        raise ValueError("Please setup the GITHUB_APIKEY env variable")
    headers = {
        "Authorization": f"token {apiKey}"
    }
    res = requests.get(url, headers=headers, params=queryParams)
    #print(res.url)
    return res.json()
```

```python
readme = getFromGithub ("/repos/ironhack-datalabs/scavenger/git/trees/master?recursive=1")
readme
```

```python
archivos = {}
for e in readme['tree']:
    if '.scavengerhunt' in e['path']:
        filename = e['path'].split('/.')
        filename2 = filename[1].split('.')
        print(filename2)
        archivos[filename2[0]] = e['path']

print(archivos)
```

```python
import base64
mensaje = ''
for e in sorted(archivos):
    mess = getFromGithub ("/repos/ironhack-datalabs/scavenger/contents"+archivos[e])
    x = base64.standard_b64decode(mess['content'])
    mensaje += str(x)
    #print(mensaje)
    #print(type(x))
    #print(base64.standard_b64decode(mess['content']))
print(mensaje)
```

```python
palabras = re.findall(r'\w+', mensaje)
joke = ''
for e in palabras:
    if e!= 'b' and e!= 'n':
        joke += e
        joke += ' '
print(joke)
```
